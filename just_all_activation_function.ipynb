# All activation functions

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import time
```

# Custom Asymmetric Activation

```python
def custom_asymmetric_activation(x):
    alpha = 0.2
    beta = 1
    k = 2
    return np.where(x < 0, -alpha * np.log(1 + np.exp(x)), beta * (1 - np.exp(-k * x)))

x = np.linspace(-5, 5, 100)
y = custom_asymmetric_activation(x)

plt.plot(x, y, color='blue', label='Custom Asymmetric Activation')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title("Custom Asymmetric Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.legend()
plt.grid(True)
plt.show()
```

# Quadratic activation

```python
# Define the Quadratic Activation Function
def quadratic_activation(x):
    a, b, c = 0.5, 0.0, 0.0  # Example coefficients
    return a * tf.square(x) + b * x + c


x = np.linspace(-5, 5, 100)
y = quadratic_activation(x)

plt.plot(x, y, color='blue', label='Quadratic Activation')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title("Quadratic Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.legend()
plt.grid(True)
plt.show()
```

# Taylor series ctivation

```python
# Define the Taylor Series Inspired Activation
def taylor_series_activation(x):
    a0, a1, a2, a3 = 0.0, 0.5, 0.1, 0.01  # Example coefficients
    return a0 + a1 * x + a2 * tf.square(x) + a3 * tf.pow(x, 3)


x = np.linspace(-5, 5, 100)
y = taylor_series_activation(x)

plt.plot(x, y, color='blue', label='Taylor series Activation')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title("Taylor series Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.legend()
plt.grid(True)
plt.show()
```

# Quantum tunneling activation

```python
# Define Quantum-Tunneling-inspired Activation Function
def quantum_tunneling_activation(x):
    k = 0.5  # Scaling factor for tunneling effect
    epsilon = 0.1  # Small bias term to simulate tunneling for smaller values
    return x * tf.keras.activations.sigmoid(k * x) + epsilon


x = np.linspace(-5, 5, 100)
y = quantum_tunneling_activation(x)

plt.plot(x, y, color='blue', label='Quantum tunneling Activation')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title("Quantum tunneling Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.legend()
plt.grid(True)
plt.show()
```

# Quantum Hyperbolic Activation

```python
# Define the Quantum Hyperbolic Activation function
def QHA(x, alpha=1.0, beta=0.1):
    return 0.5 * (np.exp(alpha * x) - np.exp(-alpha * x)) * np.sign(x) + beta * x

# Generate input data
x = np.linspace(-5, 5, 100)
y = QHA(x)

# Plotting the QHA Activation Function
plt.plot(x, y, color='blue', label='QHA Activation')
plt.title('Quantum Hyperbolic Activation Function (QHA)')
plt.xlabel('Input (x)')
plt.ylabel('Output (QHA(x))')
plt.axhline(0, color='black',linewidth=0.5, ls='--')
plt.axvline(0, color='black',linewidth=0.5, ls='--')
plt.grid(True)
plt.legend()
plt.show()
```

# Exponential Range Activation (ERA) function

```python
# Define the Exponential Range Activation (ERA) function
def ERA(x):
    # Get the integer part of x (floor)
    n = tf.floor(x)
    # Calculate the output for each x based on its interval
    return tf.cast(n, tf.float32) + (1 - tf.exp(-(x - n)))

# Generate input values for both ranges
x1 = np.linspace(-5, 5, 100)  # Range from -5 to 5
x2 = np.linspace(0, 5, 100)   # Range from 0 to 5

# Convert to TensorFlow constants
x1_tf = tf.constant(x1, dtype=tf.float32)
x2_tf = tf.constant(x2, dtype=tf.float32)

# Apply the ERA function to both sets of inputs
y1 = ERA(x1_tf).numpy()
y2 = ERA(x2_tf).numpy()

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Plot for x1 (from -5 to 5)
ax1.plot(x1, y1, color='blue', label='ERA for x = np.linspace(-5, 5, 100)')
ax1.set_title('ERA for x = np.linspace(-5, 5, 100)')
ax1.set_xlabel('Input (x)')
ax1.set_ylabel('Output (f(x))')
ax1.axhline(0, color='black', linewidth=0.5, ls='--')
ax1.axvline(0, color='black', linewidth=0.5, ls='--')
ax1.grid(True)
ax1.set_xticks(range(-5, 6, 1))  # Show integer ticks on x-axis
ax1.set_yticks(range(int(np.floor(y1.min())), int(np.ceil(y1.max())) + 1, 1))  # Integer ticks on y-axis
ax1.legend()

# Plot for x2 (from 0 to 5)
ax2.plot(x2, y2, color='red', label='ERA for x = np.linspace(0, 5, 100)')
ax2.set_title('ERA for x = np.linspace(0, 5, 100)')
ax2.set_xlabel('Input (x)')
ax2.axhline(0, color='black', linewidth=0.5, ls='--')
ax2.axvline(0, color='black', linewidth=0.5, ls='--')
ax2.grid(True)
ax2.set_xticks(range(0, 6, 1))  # Show integer ticks on x-axis
ax2.set_yticks(range(int(np.floor(y2.min())), int(np.ceil(y2.max())) + 1, 1))  # Integer ticks on y-axis
ax2.legend()

# Display the plots
plt.tight_layout()
plt.show()
```

# Smooth Exponential Range Activation (ERA_smooth) function

```python
# Define the Exponential Range Activation (ERA) function with smooth transitions
def era_smooth(x, alpha=5):
    # Get the integer part of x (floor)
    n = tf.floor(x)
    # Calculate the output for each x based on its interval with smooth transitions
    return n + (1 - tf.exp(-alpha * (x - n))) * tf.cast(x - n < 1, tf.float32)

# Generate input values
x_values1 = tf.linspace(-5.0, 5.0, 100)  # From -5 to 5 for a broader range
x_values2 = tf.linspace(0.0, 5.0, 100)   # From 0 to 5, original range

# Calculate output values
y_values1 = era_smooth(x_values1)
y_values2 = era_smooth(x_values2)

# Plot the results using subplots
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# First subplot: Smooth ERA function on range [-5, 5]
axs[0].plot(x_values1, y_values1, label='ERA Smooth [-5, 5]', color='blue')
axs[0].set_title('ERA Smooth Function on [-5, 5]')
axs[0].set_xlabel('Input (x)')
axs[0].set_ylabel('Output (f(x))')
axs[0].grid()
axs[0].axhline(0, color='black', lw=0.5, ls='--')
axs[0].axvline(0, color='black', lw=0.5, ls='--')
axs[0].set_xticks(range(-5, 6))  # Show x ticks from -5 to 5
axs[0].set_yticks(range(-5, 6))  # Show y ticks from -5 to 5
axs[0].legend()

# Second subplot: Smooth ERA function on range [0, 5]
axs[1].plot(x_values2, y_values2, label='ERA Smooth [0, 5]', color='green')
axs[1].set_title('ERA Smooth on [0, 5]')
axs[1].set_xlabel('Input (x)')
axs[1].set_ylabel('Output (f(x))')
axs[1].grid()
axs[1].axhline(0, color='black', lw=0.5, ls='--')
axs[1].axvline(0, color='black', lw=0.5, ls='--')
axs[1].set_xticks(range(0, 6))  # Show x ticks from 0 to 5
axs[1].set_yticks(range(0, 6))  # Show y ticks from 0 to 5
axs[1].legend()

plt.tight_layout()
plt.show()
```

# Step Base Activation Function

```python
# Define the step_base activation function
def step_base(x):
    # Zero for x <= 0
    positive_part = tf.maximum(x, 0)
    
    # Floor for the step base
    step_base = tf.floor(positive_part)
    
    # Smooth rise within each step
    fractional_part = 1 - tf.exp(-tf.math.mod(positive_part, 1.0))
    
    # Combine both parts
    return step_base + fractional_part


# Generate input values
x_values1 = tf.linspace(-5.0, 5.0, 100)  # From -5 to 5 for a broader range
x_values2 = tf.linspace(0.0, 5.0, 100)   # From 0 to 5, original range

# Calculate output values
y_values1 = step_base(x_values1)
y_values2 = step_base(x_values2)

# Plot the results using subplots
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# First subplot: step_base function on range [-5, 5]
axs[0].plot(x_values1, y_values1, label='step_base [-5, 5]', color='blue')
axs[0].set_title('step_base Function on [-5, 5]')
axs[0].set_xlabel('Input (x)')
axs[0].set_ylabel('Output (f(x))')
axs[0].grid()
axs[0].axhline(0, color='black', lw=0.5, ls='--')
axs[0].axvline(0, color='black', lw=0.5, ls='--')
axs[0].set_xticks(range(-5, 6))  # Show x ticks from -5 to 5
axs[0].set_yticks(range(-5, 6))  # Show y ticks from -5 to 5
axs[0].legend()

# Second subplot: step_base function on range [0, 5]
axs[1].plot(x_values2, y_values2, label='step_base [0, 5]', color='green')
axs[1].set_title('step_base on [0, 5]')
axs[1].set_xlabel('Input (x)')
axs[1].set_ylabel('Output (f(x))')
axs[1].grid()
axs[1].axhline(0, color='black', lw=0.5, ls='--')
axs[1].axvline(0, color='black', lw=0.5, ls='--')
axs[1].set_xticks(range(0, 6))  # Show x ticks from 0 to 5
axs[1].set_yticks(range(0, 6))  # Show y ticks from 0 to 5
axs[1].legend()

plt.tight_layout()
plt.show()
```

# Upward S-shaped Activation Function

```python
def upward_S_shaped_activation_function(x, k=3):
    # Define output for x <= 0
    output = tf.where(x <= 0, tf.zeros_like(x), tf.zeros_like(x))
    
    # Apply upward S-shaped transitions within each integer interval for positive x
    for n in range(0, 6):  # Assuming we're visualizing up to x = 5
        interval_start = tf.cast(n, tf.float32)
        
        # Upward exponential transition within each interval
        upward_transition = interval_start + (1 - tf.exp(-k * (x - interval_start)))
        
        # Accumulate results based on the interval
        output = tf.where((x >= interval_start) & (x < interval_start + 1), upward_transition, output)
    
    return output

# Generate input values from -3 to 5
# Generate input values
x_values1 = tf.linspace(-5.0, 5.0, 100)  # From -5 to 5 for a broader range
x_values2 = tf.linspace(0.0, 5.0, 100)   # From 0 to 5, original range

# Calculate output values
y_values1 = upward_S_shaped_activation_function(x_values1)
y_values2 = upward_S_shaped_activation_function(x_values2)

# Plot the results using subplots
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# First subplot: upward_S_shaped_activation_function on range [-5, 5]
axs[0].plot(x_values1, y_values1, label='upward_S_shaped_activation_function [-5, 5]', color='blue')
axs[0].set_title('upward_S_shaped_activation_function Function on [-5, 5]')
axs[0].set_xlabel('Input (x)')
axs[0].set_ylabel('Output (f(x))')
axs[0].grid()
axs[0].axhline(0, color='black', lw=0.5, ls='--')
axs[0].axvline(0, color='black', lw=0.5, ls='--')
axs[0].set_xticks(range(-5, 6))  # Show x ticks from -5 to 5
axs[0].set_yticks(range(-5, 6))  # Show y ticks from -5 to 5
axs[0].legend()

# Second subplot: upward_S_shaped_activation_function on range [0, 5]
axs[1].plot(x_values2, y_values2, label='upward_S_shaped_activation_function [0, 5]', color='red')
axs[1].set_title('upward_S_shaped_activation_function on [0, 5]')
axs[1].set_xlabel('Input (x)')
axs[1].set_ylabel('Output (f(x))')
axs[1].grid()
axs[1].axhline(0, color='black', lw=0.5, ls='--')
axs[1].axvline(0, color='black', lw=0.5, ls='--')
axs[1].set_xticks(range(0, 6))  # Show x ticks from 0 to 5
axs[1].set_yticks(range(0, 6))  # Show y ticks from 0 to 5
axs[1].legend()

plt.tight_layout()
plt.show()
```

# Downward S-shaped Activation Function

```python
def Downward_S_shaped_activation_function(x, k=1.5):
    # Define output for x <= 0
    output = tf.where(x <= 0, tf.zeros_like(x), tf.zeros_like(x))
    
    # Apply more pronounced S-shaped transitions within each integer interval for positive x
    for n in range(0, 6):  # Assuming we're visualizing up to x = 5
        interval_start = tf.cast(n, tf.float32)
        
        # More pronounced downward exponential transition within each interval
        downward_transition = interval_start + tf.exp(-k * (x - interval_start))
        
        # Accumulate results based on the interval
        output = tf.where((x >= interval_start) & (x < interval_start + 1), downward_transition, output)
    
    return output

# Generate input values from -3 to 5
x_values = np.linspace(-3, 5, 400)
x_tensor = tf.constant(x_values, dtype=tf.float32)

# Compute the output using the custom activation function
y_values = Downward_S_shaped_activation_function(x_tensor)

# Evaluate the function to plot
y_values_evaluated = y_values.numpy()

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values_evaluated, label='Downward_S_shaped_activation_function)', color='red', linewidth=2)
plt.title("Downward S-shaped Activation Function")
plt.xlabel("Input (x)")
plt.ylabel("Output (f(x))")
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.show()
```
